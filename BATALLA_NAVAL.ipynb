```python
import random  # Importa el m√≥dulo random para generar posiciones aleatorias.

def crear_tablero():# Esta funci√≥n genera una matriz de 5x5 llena de "üåä" representando el agua.
    return [["üåä" for _ in range(5)] for _ in range(5)]

def mostrar_tablero(tablero, ocultar=True):# Imprime el tablero en la consola. Si ocultar=True, los barcos "üö¢" se muestran como agua "üåä".
    for fila in tablero:  # -----------------Itera sobre cada fila del tablero.
        for celda in fila:  # ---------------Itera sobre cada celda dentro de la fila.
            if ocultar and celda == "üö¢": #-Si ocultar es True, oculta los barcos mostrando agua.
                print("üåä", end=" ")  # ----Imprime el agua sin salto de l√≠nea.
            else:
                print(celda, end=" ")  # Imprime el contenido de la celda sin salto de l√≠nea.
        print()  # ----------------------Salto de l√≠nea al final de cada fila.
    print()  # --------------------------Espacio adicional despu√©s de mostrar el tablero.

def colocar_barcos_aleatorio(tablero):# --Ubica 3 barcos en posiciones aleatorias en el tablero de la m√°quina.
    for _ in range(3):  # ----------------Bucle que itera 3 veces para colocar los 3 barcos.
        while True:  # -------------------Bucle infinito que se ejecuta hasta encontrar una celda vac√≠a.
            x = random.randint(0, 4)  # --Genera una coordenada aleatoria en la fila.
            y = random.randint(0, 4)  # --Genera una coordenada aleatoria en la columna.
            if tablero[x][y] == "üåä": # --Verifica si la celda est√° vac√≠a.
                tablero[x][y] = "üö¢"  # --Coloca un barco en la celda vac√≠a.
                break  # ------------------Sale del bucle una vez que el barco ha sido colocado.

def colocar_barcos_jugador(tablero):#---Funcion que permite al jugador ingresar manualmente las coordenadas de sus 3 barcos
    for i in range(3):  #---------------Repite el proceso 3 veces para colocar los 3 barcos.
        while True:  #------------------Bucle infinito que se ejecuta hasta que el jugador ingrese una posici√≥n v√°lida.
            try:
                print(f"Ingresa la posici√≥n del barco {i+1}:")  # -----------Indica qu√© barco se est√° colocando, ejemplo 1,2 o 3.
                x = int(input("Fila (0-4): "))  # ----------Solicita la fila.
                y = int(input("Columna (0-4): "))  #--------Solicita la columna.
                if 0 <= x < 5 and 0 <= y < 5 and tablero[x][y] == "üåä": #----Verifica que la entrada sea v√°lida.
                    tablero[x][y] = "üö¢"  #-----------------Coloca el barco en la celda seleccionada.
                    break  # --------------------------------Sale del bucle si la posici√≥n es v√°lida.
                else:
                    print("‚ùå Ya hay un barco en esa posici√≥n. Intenta de nuevo.")  #---Si posicion invalida, muestra mensaje de error y pide nuevamente la entrada.
            except ValueError:
                print("‚ùå Entrada inv√°lida. Ingresa n√∫meros entre 0 y 4.")  #-----------Mensaje si la entrada no es un n√∫mero.

# Turno del jugador
def turno_jugador(tablero_enemigo, tablero_ataques):#-----Funcion que permite al jugador seleccionar coordenadas para atacar el tablero enemigo.
    while True:  # ----Bucle infinito que se ejecuta, hasta que el jugador ingrese coordenadas v√°lidas.
        try:
            print("Ingresa las coordenadas para atacar:")#--Mensaje que indica al jugador que debe atacar.
            x = int(input("Fila (0-4): "))  #---------------Solicita la fila para el ataque.
            y = int(input("Columna (0-4): "))  #------------Solicita la columna para el ataque.
            if 0 <= x < 5 and 0 <= y < 5:  # ---------------Verifica que las coordenadas est√©n dentro del tablero.
                if tablero_enemigo[x][y] == "üö¢":  #-------Verifica si hay un barco en esa posici√≥n.
                    tablero_ataques[x][y] = tablero_enemigo[x][y] = "üí•" # --Marca al impactar un barco enemigo.
                    print("üî•¬°Acierto! Has hundido un barco enemigo.")  # ---Mensaje de impacto exitoso.
                else:
                    tablero_ataques[x][y] = "‚ùå"  #----Marca al impactar con el agua (fallo).
                    print("üí¶ Agua. Fallaste.") #------Mensaje de impacto fallido.
                break  #--- Sale del bucle despu√©s del ataque.
            else: #---------Si las coordenadas ingresadas estan fuera de rango
                print("‚ùå Coordenadas fuera de rango. Intenta de nuevo.") #------Muestra mensaje de error
        except ValueError: #---------------Captura un error si el usuario ingresa un dato incorrecto
            print("‚ùå Entrada inv√°lida. Ingresa n√∫meros entre 0 y 4.")

# Turno de la m√°quina
def turno_maquina(tablero_jugador):# Funcion que permite a la m√°quina seleccionar coordenadas aleatorias para atacar al jugador.
    while True:
        x = random.randint(0, 4)  # ------Genera una fila aleatoria para el ataque.
        y = random.randint(0, 4)  # ------Genera una columna aleatoria para el ataque.
        if tablero_jugador[x][y] not in ["üí•", "‚ùå"]:  #---Verifica que la posici√≥n no haya sido atacada.
            break
    print(f"La m√°quina ataca en ({x}, {y})")
    if tablero_jugador[x][y] == "üö¢":  #---------Verifica si hay un barco en esa posici√≥n.
        tablero_jugador[x][y] = "üí•"  # ---------Marca impacto exitoso.
        print("üíÄ La m√°quina acert√≥, hundio uno de tus barcos.")
    else:
        tablero_jugador[x][y] = "‚ùå"  # Marca al impactar con el agua (fallo).
        print("ü§ñ La m√°quina fall√≥.")

# Verificar si quedan barcos
def todos_hundidos(tablero):#---Retorna True si todos los barcos han sido hundidos, False en caso contrario.
    for fila in tablero:  # ----Itera sobre cada fila del tablero.
        if "üö¢" in fila:  # ---Si encuentra un barco, retorna False.
            return False
    return True  # ------------Retorna True si no quedan barcos.

# Juego principal
jugador = crear_tablero()  # --------------Crea el tablero del jugador.
maquina = crear_tablero()  # --------------Crea el tablero de la m√°quina.
ataques_jugador = crear_tablero()  # ------Crea el tablero donde se registran los ataques del jugador.

colocar_barcos_jugador(jugador)  # --------El jugador coloca sus barcos.
colocar_barcos_aleatorio(maquina)  # ------La m√°quina coloca sus barcos aleatoriamente.

while True:
    print("Tu tablero:")
    mostrar_tablero(jugador, ocultar=False) #-----Muestra el tablero del jugador en la consola con sus 3 barcos.
    print("Tus ataques:")
    mostrar_tablero(ataques_jugador)#-------------Muestra el tablero donde se registran los ataques del jugador contra la m√°quina.

    turno_jugador(maquina, ataques_jugador)
    if todos_hundidos(maquina):#-----revisa si quedan barcos üö¢ en el tablero de la m√°quina.
        print("üéâ¬°Felicidades! Has hundido todos los barcos enemigos.")# ---muestra un mensaje de victoria.
        break#-------Termina el juego

    turno_maquina(jugador)
    if todos_hundidos(jugador):#-------------------------------------Revisa si quedan barcos üö¢ en el tablero del jugador.
        print("üíÄ¬°La m√°quina ha hundido todos tus barcos!.") #-------Muestra un mensaje de derrota.
        break #-----Termina el juego.

# Mostrar tableros finales016
print("Tablero final de la m√°quina:")
mostrar_tablero(maquina, ocultar=False)
print("Tu tablero final:")
mostrar_tablero(jugador, ocultar=False)

```
